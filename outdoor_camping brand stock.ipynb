```python

import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf

outdoor_brand = ['DKS','WGO','CWH',]
df_yahoo = yf.download(outdoor_brand, start='2019-04-27', end='2021-04-27', progress=False)

df_yahoo2 = df_yahoo[[('Close', i) for i in outdoor_brand]].copy()
df_yahoo2.columns = [f'{i}_close' for i in outdoor_brand]
df_yahoo2 = df_yahoo2.reset_index()
df_yahoo2.columns

fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(12, 8))
for i in outdoor_brand:
    ax.plot(df_yahoo2['Date'], df_yahoo2[f'{i}_close'], label=i)
ax.legend()
```
